```python
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
'''
自訂函數 sign_fun() 並且作圖（自由取一段 x 區間）
'''
def sign_fun(x):
    if i > 0:
        ans = 1
    elif i < 0:
        ans = -1
    else:
        ans = 0
    return ans

x = np.linspace(-10, 10, 100)
sign_fun_y = np.zeros(100)
t = 0

for i in x:
    sign_fun_y[t] = sign_fun(i)
    t += 1
    
# 作圖
plt.plot(x, sign_fun_y)

# 顯示
plt.show()
```

```python
'''
自訂函數 sigmoid_fun() 並且作圖（自由取一段 x 區間）
'''

def sigmoid_fun(x):
    return 1/(1 + math.exp(-x))

x = np.linspace(-10, 10, 100)
sigmoid_fun_y = np.zeros(100)
t = 0

for i in x:
    sigmoid_fun_y[t] = sigmoid_fun(i)
    t += 1
    
# 作圖
plt.plot(x, sigmoid_fun_y)

# 顯示
plt.show()
```

```python
'''
承前題，將前兩個函數畫在同一個函數圖（Optional）
'''
# 作圖
plt.plot(x, sign_fun_y)
plt.plot(x, sigmoid_fun_y)

# 顯示
plt.show()
```

```python
'''
承前題，利用 subplot() 將這兩個圖畫在 (1, 2) 的畫布上（Optional）
'''
plt.subplot(1, 2, 1)
plt.plot(x, sign_fun_y)
plt.subplot(1, 2, 2)
plt.plot(x, sigmoid_fun_y)

plt.show()
```
